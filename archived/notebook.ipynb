```python
from benford import *
import matplotlib.pyplot as plt
```

```python
dir = 'data'
files = load_image_files('data')
len(files)
```

```python
images = read_images(dir, files)
len(images)
```

```python
bf = benford_pct(firsts(gray(images[0])))
bf
```

```python
# Plot the Benford distribution and image side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.set_title('Benford distribution')
ax2.imshow(images[0])
ax2.set_title('Image')

ax1.plot(range(1,10), benford_pct(firsts(gray(images[0]))), 'ko-')
ax1.plot(range(1,10), benford_pct(firsts(red(images[0]))), 'ro-')
ax1.plot(range(1,10), benford_pct(firsts(green(images[0]))), 'go-')
ax1.plot(range(1,10), benford_pct(firsts(blue(images[0]))), 'bo-')
ax1.plot(range(1,10), benford_pct(firsts(hue(images[0]))), 'co-')
ax1.plot(range(1,10), benford_pct(firsts(sat(images[0]))), 'mo-')
ax1.plot(range(1,10), benford_pct(firsts(val(images[0]))), 'yo-')
ax1.legend(['Gray', 'Red', 'Green', 'Blue', 'Hue', 'Saturation', 'Value'])
plt.show()

# Save the image 
fig.savefig('benford.png')
```
