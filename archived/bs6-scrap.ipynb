```python
import cv2 
import matplotlib.pyplot as plt 
import numpy as np
import collections
```

```python
# Read an image 
img = cv2.imread('data/gs/image_17.jpg')

# Convert image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# resize image to 10%
gray = cv2.resize(gray, (0,0), fx=0.1, fy=0.1)

# Display the grayscale image 
plt.imshow(gray, cmap='gray')
```

```python
# Convert image into 1d signal
signal = gray.flatten()

# Normalise the amplitude between -1 to 1
signal = signal / 255.0
signal = signal - 0.5
signal = signal * 2

fig = plt.figure(figsize=(25, 5))
plt.plot(signal)
```
