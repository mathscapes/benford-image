```python
import cv2 
import matplotlib.pyplot as plt 
import collections
```

```python
# Read an image 
img = cv2.imread('data/gs/image_7.jpg')

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
```

```python
# Pixel values 
print(gray)
```

```python
# Flatten the image
flat = gray.flatten()
print(flat)
```

```python
# Count the frequency of each value from 0-255
counts = collections.Counter(flat)

# Sorting the counts because collections.Counter doesn't preserve order
counts_sorted = [ counts[i] for i in range(256) ]
   
print(counts_sorted)
```

```python
# Plot the counts 
plt.plot(range(0, 256), counts_sorted)
```

```python
# Get a count of leading digits in counts_sorted 
leading = [ 0 for i in range(10) ]
for i in range(10):
    for j in range(10):
        leading[i] += counts_sorted[i * 10 + j]
```
