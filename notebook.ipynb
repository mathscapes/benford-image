```python
from benford import *
import matplotlib.pyplot as plt
```

```python
dir = 'data'
files = load_image_files('data')
len(files)
```

```python
images = read_images(dir, files)
len(images)
```

```python
bf = benford_pct(firsts(gray(images[0])))
bf
```

```python
# Plot the Benford distribution
plt.plot(range(1,10), benford_pct(firsts(gray(images[0]))), 'ko-')
plt.plot(range(1,10), benford_pct(firsts(red(images[0]))), 'ro-')
plt.plot(range(1,10), benford_pct(firsts(green(images[0]))), 'go-')
plt.plot(range(1,10), benford_pct(firsts(blue(images[0]))), 'bo-')
plt.plot(range(1,10), benford_pct(firsts(hue(images[0]))), 'co-')
plt.plot(range(1,10), benford_pct(firsts(sat(images[0]))), 'mo-')
plt.plot(range(1,10), benford_pct(firsts(val(images[0]))), 'yo-')
plt.title('Benford Distribution')
plt.xlabel('Digit')
plt.ylabel('Percentage')
plt.legend(['Gray', 'Red', 'Green', 'Blue', 'Hue', 'Saturation', 'Value'])
plt.show()
```
